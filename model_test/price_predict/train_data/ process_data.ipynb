```python
import pandas as pd
import numpy as np
from datetime import datetime,timezone
import time
```

```python
df = pd.read_csv('Ethereum_train.csv')
```

```python
def make_emotion_label(column_name):
    df['emotion'] = np.where(df[column_name] >= 0.5, 1 ,df[column_name])
    df['emotion'] = np.where((df[column_name] > -0.5) & (df[column_name] < 0.5) , 0 ,df['emotion'])
    df['emotion'] = np.where(df[column_name] <= -0.5, -1 ,df['emotion'])
```

```python
make_emotion_label("vader")
```

```python
df.drop(df.columns[0],axis=1,inplace=True)
```

```python
df
```

```python
df.sort_values(by = ['created_utc'],inplace=True)
```

```python
def utc_to_local(utc_time):
    local_time = time.gmtime(utc_time)
    return f"{local_time[0]}-{local_time[1]}-{local_time[2]}"
```

```python
df['created_utc'] = df['created_utc'].apply(utc_to_local)
```

```python
df = df.rename(columns={'created_utc':'date_time'})
```

```python
df.to_csv('./with_emotion/Ethereum.csv')
```
