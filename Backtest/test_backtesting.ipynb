```python
from backtesting import Backtest, Strategy
from backtesting.lib import crossover
from backtesting.test import SMA
import pandas as pd


class SmaCross(Strategy):
    def init(self):
        senti = self.data.Sent
        self.ma1 = self.I(SMA, senti, 5)
        self.ma2 = self.I(SMA, senti, 40)

    def next(self):
        if crossover(self.ma1, self.ma2):
            self.buy()
        elif crossover(self.ma2, self.ma1):
            self.sell()


btc = pd.read_csv("./BTC-USD.csv", index_col=0, parse_dates=True, infer_datetime_format=True)
btc_bt = Backtest(btc, SmaCross, commission=.002, exclusive_orders=True, cash=1000000)

stats = btc_bt.run()
btc_bt.plot()
```
